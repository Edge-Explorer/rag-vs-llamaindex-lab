# LlamaIndex RAG: 01 Basic Pipeline

In this notebook, we'll build a basic RAG pipeline using LlamaIndex.

```python
import os
import nest_asyncio
from dotenv import load_dotenv
from llama_index.llms.google_genai import GoogleGenAI
from llama_index.embeddings.google_genai import GoogleGenAIEmbedding
from llama_index.core import Settings, VectorStoreIndex, Document
```

```python
load_dotenv()
```

```python
nest_asyncio.apply()
```

```python
Settings.llm = GoogleGenAI(model="models/gemini-2.0-flash", api_key=os.getenv("GEMINI_API_KEY"))

Settings.embed_model = GoogleGenAIEmbedding(model_name="models/gemini-embedding-001", api_key=os.getenv("GEMINI_API_KEY"))
```

```python
text = """Retrieval-Augmented Generation (RAG) is a technique that gives LLMs access to specific, private data. 
It works by combining a retrieval system with a generative model. 
The retrieval system finds relevant documents, and the model uses them to generate an answer. 
This prevents 'hallucinations' because the model can ground its response in the provided context."""

document= Document(text= text)
```

```python
index= VectorStoreIndex.from_documents([document])
```

```python
query_engine= index.as_query_engine()
```

```python
response= query_engine.query("What is RAG?")
```

```python
print(f"LlamaIndex Answer: {response}")
```
