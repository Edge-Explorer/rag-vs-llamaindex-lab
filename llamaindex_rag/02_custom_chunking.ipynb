```python
from llama_index.core.node_parser import SentenceSplitter

splitter= SentenceSplitter(chunk_size= 50, chunk_overlap= 10)
```

```python
text= "RAG (Retrieval-Augmented Generation) is a powerful technique for AI. " * 20
doc= Document(text= text)
```

```python
nodes= splitter.get_nodes_from_documents([doc])
```

```python
print(f"Total Chunks (Nodes): {len(nodes)}")
for i, node in enumerate(nodes[:3]):
    print(f"\n--- Chunk {i} ---\n{node.get_content()}")
```

```python
custom_index = VectorStoreIndex.from_documents([doc], transformations=[splitter])

query_engine = custom_index.as_query_engine()
response = query_engine.query("What is the main technique mentioned?")
print(f"Answer with custom chunks: {response}")
```
