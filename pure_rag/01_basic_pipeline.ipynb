# Pure RAG: 01 Basic Pipeline

In this notebook, we'll build a basic RAG pipeline from scratch.

## 1. Loader

We'll load a sample document and prepare it for chunking.

```python
import os
import numpy as np
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
sample_text= """ Retrieval-Augmented Generation (RAG) is a technique that gives LLMs access to specific, private data. 
It works by combining a retrieval system with a generative model. 
The retrieval system finds relevant documents, and the model uses them to generate an answer. 
This prevents 'hallucinations' because the model can ground its response in the provided context. """

chunk= sample_text.split('. ')
```

```python
print(chunk)
```

```python
from sentence_transformers import SentenceTransformer

model= SentenceTransformer('all-miniLM-L6-v2')
```

```python
embeddings= model.encode(chunk)
```

```python
query= "What is RAG?"
query_embedding= model.encode([query])
```

```python
# Multiply our chunks by our query to find the match
similarities= np.dot(embeddings, query_embedding.T)
```

```python
best_idx= np.argmax(similarities)
print(f"Query: {query}")
print(f"Best Match (Index {best_idx}): {chunk[best_idx]}")
```
